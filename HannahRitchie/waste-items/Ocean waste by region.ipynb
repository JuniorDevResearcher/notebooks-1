```python
import pandas as pd
```

```python
waste = pd.read_csv("input/pollution_items_regions.csv", encoding="latin1")
```

```python
waste = waste.set_index(["Items", "Region"]).stack().unstack(level=1).reset_index()
```

```python
waste["Year"] = "2021"
waste = waste.drop(columns=["level_1"])
```

```python
waste
```

```python
waste = waste[ ["Year"] + [ col for col in waste.columns if col != "Year" ] ]
waste = waste[ ["Items"] + [ col for col in waste.columns if col != "Items" ] ]
```

```python
waste = waste.rename(columns={"Items":"Entity"})
```

```python
waste.to_csv("output/ocean_waste_items_by_region.csv", index=False)
```
