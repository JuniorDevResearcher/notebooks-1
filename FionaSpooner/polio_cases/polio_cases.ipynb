```python
import requests
from datetime import date, timedelta
import pdfplumber
import pandas as pd
```

```python
def download_latest_wild_polio(days_to_sub:int):
    day = date.today() - timedelta(days = days_to_sub)
    wild_polio_date = day.strftime("%Y%m%d")
    url = f"http://polioeradication.org/wp-content/uploads/2022/03/weekly-polio-analyses-WPV-{wild_polio_date}.pdf"
    res = requests.get(url)
    return(res)
```

Find the latest file by trying to download a file from today, today -1, today-2, until a file exists (res.ok = True).

```python
i = 0
download_available = False
while not download_available:    
    res = download_latest_wild_polio(days_to_sub = i)
    if res.ok:
        download_available = True
    else:
        i +=1

with open('data/polio_wild_cases.pdf', 'wb') as f:
    f.write(res.content)
```
