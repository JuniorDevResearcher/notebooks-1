```python
df = pd.read_csv('https://raw.githubusercontent.com/owid/notebooks/main/JoeHasell/Povcal_data_work/Poverty_and_inequality_measures_from_PovCal_2021.csv')
```

```python
# Check population – note that Suriname has population less than one (one person...)

df.loc[df.Population < 1][["Entity", "Year", "Population"]]
```

```python
# -> drop weird population figures....
df = df.loc[df.Population > 1]
```

```python
df_2017 = df.loc[df.Year == 2017]
```

```python
df_2017.head()
```

```python
exclude_list = ['China (Rural)','China (Urban)','East Asia and Pacific','Europe and Central Asia','India (Rural)','India (Urban)','Indonesia (Rural)','Indonesia (Urban)','Latin America and the Caribbean','Middle East and North Africa','Other high Income (World Bank)','South Asia','Sub-Saharan Africa','Suriname (Urban)','World']
```

```python
# Check population
```

```python
df_2017_countries = df_2017[~df_2017["Entity"].isin(exclude_list)]
```

```python
df_2017_countries.head()
```

```python
df_2017_countries["threshold Above $30 per day"] = df_2017_countries["Population"] - df_2017_countries["$30.00 per day - total number of people below poverty line"]
```

```python
df_2017_countries["threshold $10 - $30 per day"] = df_2017_countries["$30.00 per day - total number of people below poverty line"] - df_2017_countries["$10.00 per day - total number of people below poverty line"]
```

```python
df_2017_countries["threshold $1.90 - $10 per day"] = df_2017_countries["$10.00 per day - total number of people below poverty line"] - df_2017_countries["$1.90 per day - total number of people below poverty line"]
```

```python
df_2017_countries["threshold below $1.90 per day"] = df_2017_countries["$1.90 per day - total number of people below poverty line"]
```

```python
# Calc shares

df_2017_countries["share below $1.90 per day"] = df_2017_countries["threshold below $1.90 per day"]/df_2017_countries["Population"]
df_2017_countries["share $1.90 - $10 per day"] = df_2017_countries["threshold $1.90 - $10 per day"]/df_2017_countries["Population"]
df_2017_countries["share $10 - $30 per day"] = df_2017_countries["threshold $10 - $30 per day"]/df_2017_countries["Population"]
df_2017_countries["share Above $30 per day"] = df_2017_countries["threshold Above $30 per day"]/df_2017_countries["Population"]
df_2017_countries.head(n=20)
```

```python
df_2017_countries = df_2017_countries[['Entity',\
                                       'threshold below $1.90 per day',\
                                     "threshold $1.90 - $10 per day",\
                                       "threshold $10 - $30 per day",\
                                       "threshold Above $30 per day",\
                                      "share below $1.90 per day",\
                                      "share $1.90 - $10 per day",\
                                      "share $10 - $30 per day",\
                                      "share Above $30 per day",
                                      "Population"]]
```

```python
df_2017_countries.head()
```

```python
df_2017_countries = df_2017_countries.sort_values(by=['share Above $30 per day'])
df_2017_countries["cum_pop"] = df_2017_countries.Population.cumsum()

df_2017_countries.head()
```

```python
df_2017_countries.to_csv("povcal_number_at_different_thresholds_2007.csv")
```

```python
df_2017_countries_long = df_2017_countries[['Entity',\
                                       'threshold below $1.90 per day',\
                                     "threshold $1.90 - $10 per day",\
                                       "threshold $10 - $30 per day",\
                                       "threshold Above $30 per day",]]

df_2017_countries_long = pd.melt(df_2017_countries,id_vars=['Entity'],var_name='metric', value_name='value')
```

```python
df_2017_countries_long.to_csv("long_povcal_number_at_different_thresholds_2007.csv")
```
