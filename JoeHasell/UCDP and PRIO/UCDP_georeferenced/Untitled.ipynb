```python
import pandas as pd
import requests, zipfile, io
from zipfile import ZipFile
import gzip
```

###Load in the main UCDP georeferenced data (1989-2020)

```python
xy_file = ZipFile("data/input/ged211.csv.zip")

df_ged = pd.read_csv(xy_file.open('ged211.csv'))
df_ged.tail()
```

Note that neither `id` nor `relid` are duplicated within this dataset

```python
any(df_ged[['id']].duplicated())
```

```python
any(df_ged[['relid']].duplicated())
```

###Load in the UCDP Candidate data for 2021

```python
xy_file = ZipFile("data/input/GEDEvent_v21_01_21_12.csv.zip")

df_candidate_annual = pd.read_csv(xy_file.open('GEDEvent_v21_01_21_12.csv'))
df_candidate_annual.tail()
```

Note that neither `id` nor `relid` are duplicated within this dataset

```python
any(df_candidate_annual[['id']].duplicated())
```

```python
any(df_candidate_annual[['relid']].duplicated())
```

### Load in the UCDP Candidate data – monthly for 2022 (up to Feb only currently)

```python
xy_file = ZipFile("data/input/GEDEvent_v22_0_2.csv.zip")

df_candidate_monthly = pd.read_csv(xy_file.open('GEDEvent_v22_0_2.csv'))
df_candidate_monthly.tail()
```

Note that neither `id` nor `relid` are duplicated within this dataset

```python
any(df_candidate_monthly[['id']].duplicated())
```

```python
any(df_candidate_monthly[['relid']].duplicated())
```

### Append the GED and Candidate datasets into a single dataframe

```python
frames = [df_ged, df_candidate_annual, df_candidate_monthly]

df = pd.concat(frames)
```

Note that `id` is not unique across the datasets – there is a duplicate for Eritrea.

```python
df[df[['id']].duplicated(keep=False)]
```

But `relid` is unique across both datasets

```python
df[df[['relid']].duplicated(keep=False)]
```
