```python
import pandas as pd
import requests, zipfile, io
from zipfile import ZipFile
```

Load in the main UCDP georeferenced data (1989-2020)

```python
xy_file = ZipFile("data/input/ged211.csv.zip")

df_ged = pd.read_csv(xy_file.open('ged211.csv'))

df_ged.tail()
```

Load in the UCDP Candidate data for 2021

```python
xy_file = ZipFile("data/input/GEDEvent_v21_01_21_12.csv.zip")

df_candidate_annual = pd.read_csv(xy_file.open('GEDEvent_v21_01_21_12.csv'))
df_candidate_annual.tail()
```

Load in the UCDP Candidate data â€“ monthly for 2022 (up to Feb only currently)

```python
xy_file = ZipFile("data/input/GEDEvent_v22_0_2.csv.zip")

df_candidate_monthly = pd.read_csv(xy_file.open('GEDEvent_v22_0_2.csv'))
df_candidate_monthly.tail()
```

### Append the GED and Candidate datasets into a single dataframe

```python
frames = [df_ged, df_candidate_annual, df_candidate_monthly]

df = pd.concat(frames, keys=['ged', 'candidate_annual', 'candidate_monthly']).reset_index()
df = df.rename(columns={"level_0": "data_source"}).drop(columns=['level_1'])
```

```python
df.head()
```

### Sense check 

Note that `id` is not unique across the datasets â€“ there is a duplicate for Eritrea.

```python
df[df[['id']].duplicated(keep=False)]
```

But `relid` is unique across datasets

```python
df[df[['relid']].duplicated(keep=False)]
```

Note that there are some rows where the deaths estimates dont make sense.

... where the `high` estimate is lower than the `low` estimate:

```python
df[df['high']<df['low']].head()
```

I make an amended set of deaths data, where `high` estimates are swapped for `low` where `high`<`low`.

```python
df['high_amend'] = df['high']
```

```python
df.loc[(df.high<df.low), 'high_amend'] = df.loc[(df.high<df.low), 'low'] 
```

```python
df['low_amend'] = df['low']
```

```python
df.loc[(df.high<df.low), 'low_amend'] = df.loc[(df.high<df.low), 'high'] 
```

There are also rows where the `best` estimate falls outside the `low`-`high` range.

```python
df[(df['best']<df['low_amend']) | (df['best']>df['high_amend'])].head()
```

I extend the (amended) `high` or `low` values to include the `best` estimate.

```python
df.loc[(df.best<df.low_amend), 'low_amend'] = df.loc[(df.best<df.low_amend), 'best'] 
```

```python
df.loc[(df.best>df.high_amend), 'high_amend'] = df.loc[(df.best>df.high_amend), 'best'] 
```
